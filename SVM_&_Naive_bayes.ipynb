# Theoretical Questions

1. What is a Support Vector Machine (SVM)?

-> SVM is a supervised machine learning algorithm used for both classification and regression.

It works by finding the optimal hyperplane that best separates data points of different classes in the feature space.

The optimal hyperplane is the one with the maximum margin — the largest distance between the hyperplane and the nearest data points from each class (called support vectors).

2. What is the difference between Hard Margin and Soft Margin SVM?

-> **Hard Margin**:

*   Assumes perfect linear separation between classes (no misclassification allowed).
*   Works only on linearly separable data.

**Soft Margin**

* Allows some misclassification, introducing penalty parameter C.
* Works well with noisy, non-separable data.

3. What is the mathematical intuition behind SVM?

-> At its core, the intuition behind Support Vector Machines (SVM) is geometric. While many algorithms try to find a line that separates data, SVM tries to find the best line.

To SVM, the "best" line is the one that separates the classes with the maximum safety margin.

4. What is the role of Lagrange Multipliers in SVM?

-> They are used to convert the constrained optimization problem into a dual form, making it easier to solve using optimization techniques.

They allow computation of the optimal weight vector w in terms of the support vectors only.

5. What are Support Vectors in SVM?

-> Support vectors are the data points closest to the decision boundary (hyperplane).

They are the most important elements of the training set because they determine the position and orientation of the hyperplane.

6. What is a Support Vector Classifier (SVC)?

-> An SVC is an implementation of SVM used for classification tasks.
It seeks a decision boundary that best separates data into different classes using linear or non-linear kernels.

7. What is a Support Vector Regressor (SVR)?

-> SVR applies SVM principles to regression.
It attempts to fit a function that deviates from actual targets by a value no greater than ε (epsilon), while keeping the model as flat as possible.

8. What is the Kernel Trick in SVM?

-> The kernel trick maps input data into a higher-dimensional feature space without explicitly computing the transformation.
This allows SVM to handle non-linear classification problems efficiently.
Common kernels:

* Linear

* Polynomial

* Radial Basis Function (RBF)

9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.

-> **Linear Kernel**

This is the simplest kernel. It implies that the data is linearly separable (can be separated by a straight line or flat plane). It does not map the data into a higher dimension; it performs the calculation in the original space.

 **Polynomial Kernel**

This kernel allows the SVM to learn non-linear models. It maps the data into a higher-dimensional space defined by polynomials of the original variables, allowing for curved decision boundaries.

 **RBF (Radial Basis Function) Kernel**

Also known as the Gaussian Kernel, this is the most popular "default" kernel. It projects data into an infinite-dimensional space. It measures similarity based on the distance between two points.

10. What is the effect of the C parameter in SVM?

-> C (regularization) controls the tradeoff between maximizing the margin and minimizing classification errors.

* Large C → low bias, high variance (overfitting risk).

* Small C → high bias, low variance (underfitting risk).

11. What is the role of the Gamma parameter in RBF Kernel SVM?

-> Gamma defines how far the influence of a single training example reaches.

Low gamma: broader influence (smooth decision boundary).

High gamma: narrow influence (tight boundaries → overfitting).

12. What is the Naïve Bayes classifier, and why is it called "Naïve"4?

-> The Naive Bayes Classifier is a supervised machine learning algorithm based on Bayes' Theorem. It is used to solve classification problems by predicting the probability that a given data point belongs to a particular class.

The algorithm is called "Naive" because it makes a massive, often unrealistic assumption about your data: Conditional Independence.

13. What is Bayes’ Theorem?

-> Bayes’ Theorem is a mathematical formula used to calculate the probability of an event based on prior knowledge of conditions that might be related to that event. It is the fundamental equation for "updating your beliefs" in the face of new evidence.

14. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes.

-> **Gaussian Naïve Bayes**

This variant is used when your features are continuous values (real numbers like 3.14, -0.01, 98.6).3Assumption: It assumes that the continuous values associated with each class are distributed according to a Normal (Gaussian) Distribution (the "Bell Curve").4The Math: It uses the mean (5$\mu$) and standard deviation (6$\sigma$) of your training data to calculate the probability of a specific value occurring.7Best Use Case: Generic classification problems where features are physical measurements.8Example: Classifying Iris flowers based on sepal_width (cm) and petal_length (cm).

**Multinomial Naïve Bayes**

This variant is used when your features are discrete counts or frequencies.

Assumption: It assumes the data follows a Multinomial Distribution. It cares about the "count" or "magnitude" of a feature.


The Math: It calculates probabilities based on how often a specific event (word) occurs out of the total number of events.

Best Use Case: Text Classification (Natural Language Processing) where the "frequency" of a word matters.

Example: Classifying a news article as "Sports" vs. "Politics."

**Bernoulli Naïve Bayes**

This variant is used when your features are binary (0 or 1, Yes or No, True or False).

Assumption: It assumes the data follows a Bernoulli Distribution (like a coin flip). It only cares if a feature is present or absent.


The Math: It explicitly penalizes the non-occurrence of a feature (which Multinomial often ignores).

Best Use Case: Text Classification where the documents are short, or only the presence of a specific keyword matters, not how many times it was said.

Example: Spam Detection.

15. When should you use Gaussian Naïve Bayes over other variants?

-> Use when features are continuous and approximately follow a normal (Gaussian) distribution.

16. What are the key assumptions made by Naïve Bayes?

* Features are conditionally independent given the class label.

* No hidden correlation between predictors.

17. What are the advantages and disadvantages of Naïve Bayes?

-> **Advantages:**

* Fast and simple to implement.

* Works well on small data.

* Performs well for text classification.

**Disadvantages:**

* Strong independence assumption.

* Poor at handling correlated features.

18. Why is Naïve Bayes a good choice for text classification?

* Word occurrences are treated as independent features.

* Efficient computation using term frequencies.

* Performs well even with high-dimensional sparse data.

19. Compare SVM and Naïve Bayes for classification tasks.

-> **Naïve Bayes**

* Extremely Fast. It just needs to do a single pass over the data to count frequencies.

* Good for baselines, but often lower than SVM on complex tasks.

* Robust. It deals with probabilities, so one weird data point rarely shifts the global probability enough to break the model.

* Can handle missing data well (it just ignores that probability term).

* High. You can see exactly which words or features contributed to the probability score.


**Support Vector Machine (SVM)**

* Slow. It solves a complex quadratic optimization problem. Training time increases roughly quadratically with dataset size.

* Generally provides higher accuracy, especially on complex, non-linear datasets.

* Sensitive. SVM cares about the "Support Vectors" (the points closest to the line). If an outlier is close to the boundary, it can shift the entire decision line.

* Cannot handle missing data; you must impute or remove it first.

* Low. The model is a "black box" defined by high-dimensional vectors. It's hard to explain why it chose a class.

20. How does Laplace Smoothing help in Naïve Bayes?

-> Laplace Smoothing (also called Additive Smoothing) is a technique used to solve the "Zero Probability" (or Zero Frequency) problem in Naive Bayes classification.

Without it, the Naive Bayes algorithm is extremely brittle: if it encounters a single word in the test data that it has never seen before in the training data, the entire model crashes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.metrics import (accuracy_score, confusion_matrix, precision_score, recall_score,
                             f1_score, roc_auc_score, log_loss, mean_squared_error, mean_absolute_error,
                             precision_recall_curve)
from sklearn.preprocessing import StandardScaler
sns.set(style="whitegrid")
```

```python
# 1. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy.
# Common imports used by most tasks
from sklearn import datasets
from sklearn.svm import SVC

iris = datasets.load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

clf = SVC(kernel='rbf', C=1.0, gamma='scale', probability=False)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# 2. Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies
from sklearn.datasets import load_wine

X, y = load_wine(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

for kernel in ['linear','rbf']:
    model = SVC(kernel=kernel, gamma='scale', C=1.0)
    model.fit(X_train, y_train)
    print(f"{kernel} accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
# 3. Write a Python program to train an SVM Regressor (SVR) on a housing dataset and evaluate it using Mean Squared Error (MSE).
from sklearn.svm import SVR
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error

data = fetch_california_housing()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

svr = SVR(kernel='rbf', C=10.0, gamma='scale', epsilon=0.1)
svr.fit(X_train_s, y_train)
y_pred = svr.predict(X_test_s)
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
# 4. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary.
from sklearn.datasets import make_circles
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt

X, y = make_circles(n_samples=300, factor=0.5, noise=0.07, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

clf = SVC(kernel='poly', degree=3, C=1.0, gamma='scale')
clf.fit(X_train, y_train)

# Plot decision boundary
xx, yy = np.meshgrid(np.linspace(X[:,0].min()-0.5, X[:,0].max()+0.5, 500),
                     np.linspace(X[:,1].min()-0.5, X[:,1].max()+0.5, 500))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X[:,0], X[:,1], c=y, s=15, cmap=plt.cm.coolwarm, edgecolor='k')
plt.title("Polynomial SVM decision boundary")
plt.show()
```

```python
# 6.  Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20 Newsgroups dataset.
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

train = fetch_20newsgroups(subset='train', shuffle=True, random_state=42)
test  = fetch_20newsgroups(subset='test', shuffle=True, random_state=42)

model = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', MultinomialNB(alpha=1.0))
])
model.fit(train.data, train.target)
pred = model.predict(test.data)
print("Accuracy:", accuracy_score(test.target, pred))
```

```python
# 7. Write a Python program to train an SVM Classifier with different C values and compare the decision
# boundaries visually.
from sklearn.datasets import make_moons
from sklearn.svm import SVC
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=300, noise=0.2, random_state=42)
Cs = [0.01, 0.1, 1, 10]
fig, axes = plt.subplots(1, len(Cs), figsize=(16,4))
for ax, C in zip(axes, Cs):
    clf = SVC(kernel='rbf', C=C, gamma='scale')
    clf.fit(X, y)
    xx, yy = np.meshgrid(np.linspace(X[:,0].min()-0.5, X[:,0].max()+0.5, 300),
                         np.linspace(X[:,1].min()-0.5, X[:,1].max()+0.5, 300))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3)
    ax.scatter(X[:,0], X[:,1], c=y, s=12, edgecolor='k')
    ax.set_title(f"C={C}")
plt.show()
```

```python
# 26. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline

import matplotlib.pyplot as plt
import numpy as np

# Ensure X_test and y_test are from the Breast Cancer dataset for consistency
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Train a new SVM model (svm_model) for this task to avoid global variable conflicts
svm_model = make_pipeline(StandardScaler(), SVC(kernel='rbf', random_state=42))
svm_model.fit(X_train, y_train)

# For SVC, decision_function gives distance to hyperplane, which is suitable for PR curve
probs = svm_model.decision_function(X_test)
precision, recall, thresholds = precision_recall_curve(y_test, probs)
avg_prec = average_precision_score(y_test, probs)

plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label=f'AP={avg_prec:.2f}')
plt.xlabel('Recall'); plt.ylabel('Precision'); plt.title('Precision-Recall curve'); plt.legend(); plt.show()
```
